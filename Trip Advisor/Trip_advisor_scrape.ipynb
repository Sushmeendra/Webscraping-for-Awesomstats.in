```python
from bs4 import BeautifulSoup
from datetime import datetime
from calendar import monthrange
import csv
import glob
import pandas as pd
    
#generates a list of filenames of .html extension in current dir        
files = glob.glob('/Users/manaranjan/Documents/Work/MyLearnings/test/tripadvisor/review_500/in/304554/301778/*.html')
```

```python
files[0:10]
```

```python
all_titles = []
all_comments = []
all_review_dates = []
all_stay_dates = []

for file in files:
    with open(file,'r+') as f:
        soup = BeautifulSoup(f, 'lxml')                            
        comment_section = soup.findAll('div',
                                        attrs = {'class': 'prw_rup prw_reviews_review_resp'})  
               
        for each_section in comment_section:       
            #comment title
            find_title = each_section.find('span',
                                      attrs = {'class': 'noQuotes'})
            all_titles.append(find_title.text)

            #comment text
            find_comment = each_section.find('div',
                                        attrs = {'class': 'entry'})
            all_comments.append(find_comment.text)

            #comment date
            find_date = each_section.find('span',
                                     attrs = {'class': 'ratingDate'})
            all_review_dates.append(find_date.text.replace('Reviewed ',''))
```

```python
reviews_df = pd.DataFrame( data = {'title' : all_titles,
                           'comment': all_comments,
                           'review_date': all_review_dates} )
```
